### Listas
En Python, una lista es una estructura de datos integrada que puede almacenar una colección ordenada de elementos. A diferencia de los arrays en otros lenguajes, las listas de Python son muy flexibles:

* Pueden contener elementos duplicados.
* Son mutables: los elementos se pueden modificar, reemplazar o eliminar.
* Son ordenadas: mantienen el orden en que se agregan los elementos.
* Se accede a los elementos por índice (desde 0).
* Pueden almacenar tipos de datos mixtos (enteros (int), cadenas (str), booleanos (bool) e incluso otras listas (list)).

___
#### Definición de una lista
Formas de definir una lista:
1. Con los paréntesis cuadrados:

````python
nums = [0, 1, 2, 3, 4] # List of integers
frutas = ['apple', 'banana', 'cherry'] # List of strings
mix = [1, 'hello', 3.14, True] # Mixed data types
print(nums)
print(frutas)
print(mix)
````

2. Con el constructor de la clase lista:

````python
a = list((1, 2, 3, 'apple', 4.5))  
b = list("GFG")
print(a)
print(b)
````

3. Repitiendo el elemento

````python
a = [2] * 5
b = [0] * 7
c = ['Hola'] * 3
print(a)
print(b)
print(c)
````

#### Función ```len(<list>)```
Función que nos permite contar los elementos de una lista. Retorna el largo de una lista


````python
print("nums:",nums)
print("Largo de num:",len(nums))
print("frutas:",frutas)
print("Largo de frutas:",len(frutas))
print("mix:",mix)
print("Largo de mix:",len(mix))
````
