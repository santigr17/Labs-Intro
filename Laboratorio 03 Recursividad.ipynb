Una función recursiva consta de dos partes clave:
* Caso base: La condición de parada que impide la recursión infinita.
* Caso recursivo: La parte de la función donde se llama a sí misma con parámetros modificados.

#### Ejemplo 1: Cálculo del factorial
Este código define una función recursiva para calcular el factorial de un número, donde la función se llama repetidamente a sí misma con valores más pequeños hasta llegar al caso base.

Aplicaciones de la función factorial: https://www.geeksforgeeks.org/maths/factorial/

```python
def factorial(n):
    if n == 0:  # Base case
        return 1
    else:       # Recursive case
        return n * factorial(n - 1)

print(factorial(5))
```

Explicación:
* Caso base: Cuando n == 0, la recursión se detiene y devuelve 1.
* Caso recursivo: Multiplica n por el factorial de n-1 hasta alcanzar el caso base.

#### Ejemplo 2: Secuencia de Fibonacci
Este código define una función recursiva para calcular el enésimo número de Fibonacci, donde cada número es la suma de los dos anteriores, comenzando desde 0 y 1.

Algunas aplicaciones de la función: https://www.geeksforgeeks.org/maths/real-life-applications-of-fibonacci-sequence/

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)

print(fibonacci(10))
```

### Tipos de recursividad

La recursión se puede clasificar en dos tipos principales: recursión de cola y recursión no de cola. La principal diferencia radica en lo que sucede después de la llamada recursiva.

Recursión de cola: La llamada recursiva es la última operación de la función, por lo que no ocurre nada después de que finaliza. Algunos lenguajes pueden optimizarla para que funcione como un bucle, ahorrando memoria.

Recursión no de cola: La función realiza más operaciones después de que finaliza la llamada recursiva, por lo que no se puede optimizar como un bucle.

Este código compara la recursión de cola y la recursión no de cola utilizando dos versiones de la función factorial: una con un acumulador (recursiva de cola) y otra con multiplicación después de la llamada recursiva (no recursiva de cola).
